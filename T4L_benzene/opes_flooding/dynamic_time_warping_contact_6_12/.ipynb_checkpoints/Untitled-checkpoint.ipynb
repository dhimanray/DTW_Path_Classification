```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = np.loadtxt('PathA/a1_a2_scaled.dat')

x = np.arange(1,len(l)+1)

plt.bar(x,l)

plt.xlabel('Residue number',fontsize=16)
plt.ylabel('DTW Distance',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0,75)
plt.text(75,36,'Lys83',fontsize=14)
plt.text(110,60,'Arg119',fontsize=14)
plt.savefig('a1_a2_dtw.png',dpi=200,bbox_inches='tight')
```
